```python
import pandas as pd
from string import  punctuation
import unicodedata
```

```python
# pd.read_sql_table('/home/arcolife/Desktop/test_data.txt')
# data = pd.read_table('/home/arcolife/Desktop/test_data.txt')
f = open('/home/arcolife/Desktop/test_data.txt', 'rb')
raw_data = f.read()
f.close()
raw_data = raw_data.splitlines()

raw_data.pop(0)
raw_data.pop(1)
raw_data.pop(-1)

# print raw_data

def clean_columns(columns):
    while '' in columns:
        columns.remove('')
    for i in xrange(len(columns)):
        columns[i] = columns[i].lstrip().rstrip()
    return columns

columns = clean_columns(raw_data.pop(0).split('|'))

def clean_data(raw_data, cleaned=True):
    for i in xrange(len(raw_data)):
        raw_data[i] = raw_data[i].split('|')
        while '' in raw_data[i]:
            raw_data[i].remove('')
        for j in xrange(len(raw_data[i])):
            raw_data[i][j] = raw_data[i][j].lstrip().rstrip()
            if cleaned:
                if j == 1 or j == 4:
                    raw_data[i][j] = raw_data[i][j].split()
                    for k in xrange(len(raw_data[i][j])):
                        raw_data[i][j][k] = raw_data[i][j][k].strip(punctuation)
                    raw_data[i][j] = [word.lower() for word in raw_data[i][j] if word]
    return raw_data

raw_data = clean_data(raw_data, cleaned=True)
data = pd.DataFrame(raw_data, columns=columns)
data
```
